### This is some code to make nice HTML tables from SQLite queries.

```python
def sqlcursor2html(c):
    colnames = map(lambda x: x[0], c.description)
    ncols = len(colnames)
    htmlstring = "<table>\n"
    htmlstring += "  <tr>"
    for name in colnames:
        htmlstring += "<th>%s</th>" % name
    htmlstring += "</tr>\n"
    for row in c:
        htmlstring += "  </tr>"
        for icol in range(ncols):
            if isinstance(row[icol], unicode):
                htmlstring += "<td>" + row[icol].encode('utf-8') + "</td>"
            else:
                htmlstring += "<td>" + str(row[icol]) + "</td>"
        htmlstring += "</tr>\n"
    htmlstring += "</table>"
    return HTML(htmlstring)

# This doesn work right now
#ip = get_ipython()
#latex_formatter = ip.display_formatter.formatters['text/html']
#latex_formatter.for_type_by_name('sqlite3', 'Cursor', sqlcursor2html)
```

## The code for building the database

```python
urlre = re.compile(r'https?://[^\s]+')
def get_one_and_two_grams(utfstring):
    """Get 1- and 2-grams from the utf string, returning a tuple of two lists.
     - URLs are removed (not used for now),
     - the UTF is mapped to an ascii subset,
     - punctuation is removed, and
     - finally a split is done on whitespace.
    """
    urls = urlre.findall(utfstring)
    utfstring = urlre.sub('', utfstring)
    norm = unidecode(utfstring)                 # Simplify the string to ascii
    norm = re.sub(r'[^a-zA-Z]', ' ', norm)      # Remove many punctuation characters
    norm = norm.lower()                         # Translate to lower case
    words = norm.split()
    digrams = map(lambda x: '%s %s' % x, zip(words[:-1],words[1:]))
    #words.extend(urls)
    return (words, digrams)

def to_datetime(datestring):
    """Turns a twitter time string into a datetime object
    I got this from
    http://stackoverflow.com/questions/7703865/going-from-twitter-date-to-python-datetime-date
    """
    time_tuple = parsedate_tz(datestring.strip())
    dt = datetime(*time_tuple[:6])
    return dt - timedelta(seconds=time_tuple[-1])

test_time_string = 'Sun Jul 21 02:08:19 +0000 2013'
test_datetime    = to_datetime(test_time_string)
test_isoformat   = test_datetime.isoformat(' ')
print "Date test:", test_time_string, "->", test_datetime, "->", test_isoformat

def build_db(limit=0, do_ngrams=True, src=None, dbfilename='ice_cream_tweets.sqlite'):
    """This will erase the current database and create a new one.
    limit= sets the maximum number of tweets saved to the database, 0 or negative means all.
    do_ngrams= can disable the creation of an ngram table which is slow.
    """
    !rm -f $dbfilename
    conn = sqlite3.connect(dbfilename)
    c = conn.cursor()
    c.execute('''CREATE TABLE tweets (tweet_id integer PRIMARY KEY, time text, tweet text, user_id integer,
                                      geolat float, geolon float, geoplace string, geocountry)''')
    if do_ngrams:
        c.execute('''CREATE TABLE ngrams (word text, tweet_id integer, n integer)''')
    #c.execute('''CREATE TABLE users (user_id integer, name text, screen_name text)''')
    ntweets = 0
    missing_tweets = 0
    duplicated_tweets = 0
    tweetids = {}
    filenames = map(lambda n: "ice_cream_tweets_%d.txt.gz" % n, range(1,4))
    print "Reading from"
    for filename in filenames:
        print "   ", filename
    if src:
        filenames = src
    for filename in filenames:
        with gzip.open(filename) as f:
            for line in f:
                if limit > 0 and ntweets >= limit:
                    break
                tweet = json.loads(line)
                if 'limit' in tweet and 'track' in tweet['limit']:
                    missing_tweets += tweet['limit']['track']
                elif 'id' in tweet and 'user' in tweet:
                    geolat = None
                    geolon = None
                    geoplace = None
                    geocountry = None
                    if 'geo' in tweet and tweet['geo']:
                        geolat,geolon = tweet['geo']['coordinates']
                    if 'place' in tweet and tweet['place']:
                        geoplace = tweet['place']['full_name']
                        geocountry = tweet['place']['country']
                    text = tweet['text']
                    tweet_id = tweet['id']
                    if tweet_id in tweetids:
                        duplicated_tweets += 1
                        #print "DUPLICATED id=%s tweet=%s" % (tweet_id, text.encode('utf-8'))
                        continue
                    tweetids[tweet_id] = 1
                    timestr = to_datetime(tweet['created_at']).isoformat(' ')
                    c.execute('INSERT INTO tweets VALUES (?, ?, ?, ?, ?, ?, ?, ?)',
                              (tweet_id, timestr, text, tweet['user']['id'],
                               geolat, geolon, geoplace, geocountry))
                    if do_ngrams:
                        words,digrams = get_one_and_two_grams(text)
                        for word in words:
                            c.execute('INSERT INTO ngrams VALUES (?, ?, ?)', (word, tweet_id, 1))
                        for digram in digrams:
                            c.execute('INSERT INTO ngrams VALUES (?, ?, ?)', (digram, tweet_id, 2))
                    ntweets += 1
                if ntweets % 1000 == 0:
                    print '.',
    conn.commit()
    conn.close()
    print " Loaded %d tweets." % ntweets
    if missing_tweets > 0:
        print "Missed %d tweets." % missing_tweets
    if duplicated_tweets > 0:
        print "%d duplicated tweets dropped." % duplicated_tweets
     
def store_sentiment(dbfile='ice_cream_tweets.sqlite'):
    conn = sqlite3.connect(dbfile)
    c = conn.cursor()
    c.execute('''DROP TABLE IF EXISTS sentiments''')
    conn.commit()
    c.execute('''CREATE TABLE sentiments (word text PRIMARY KEY, sentiment integer)''')
    with open('AFINN-111.txt') as f:
        sen_pair_reader = csv.reader(f, delimiter='\t')
        for i, word_sen in enumerate(sen_pair_reader):
            if i < 10000:
                #print word_sen
                try:
                    w = word_sen[0].decode('ascii')
                    c.execute('''INSERT INTO sentiments VALUES (?, ?)''', word_sen)
                except UnicodeDecodeError:
                    print "Sentiment table: dropping non-ascii", word_sen
                    pass
        #c.executemany('''INSERT INTO sentiments VALUES (?, ?)''', sen_pair_reader)
    conn.commit()
    conn.close()
    print "Sentiment table loaded."
   
```

## Compute sentiment of different types of tweets

This will select tweets with a certain word (say 'vanilla') and compute their total sentiment using file created by Finn Årup Nielsen in 2009-2011 as used in the Bill Howe's Data Science coursera.

```python
cmd = """SELECT time, tweet, total(sentiments.sentiment) as total_sentiment
         FROM (SELECT x.time, x.tweet_id, x.tweet, ngrams.word
         FROM (
           SELECT DISTINCT time, tweets.tweet_id, tweets.tweet
           FROM tweets,ngrams USING (tweet_id)
           WHERE ngrams.word == ?
           LIMIT ?
         ) AS x LEFT JOIN ngrams ON x.tweet_id == ngrams.tweet_id
         WHERE ngrams.n == ?
         ) AS y LEFT JOIN sentiments ON y.word == sentiments.word
         GROUP BY tweet_id
      """
sqlcursor2html(c.execute(cmd, ('vanilla', 10, 1)))
```

```python
vanilla      = np.array(map(lambda r: r[2], c.execute(cmd, ('vanilla',      100000, 1))))
chocolate    = np.array(map(lambda r: r[2], c.execute(cmd, ('chocolate',    100000, 1))))
love         = np.array(map(lambda r: r[2], c.execute(cmd, ('love',         100000, 1))))
national_ice = np.array(map(lambda r: r[2], c.execute(cmd, ('national ice', 100000, 2))))

print "vanilla        mean=%.3f stddev=%.3f" % (vanilla.mean(),      vanilla.std())
print "chocolate      mean=%.3f stddev=%.3f" % (chocolate.mean(),    chocolate.std())
print "love           mean=%.3f stddev=%.3f" % (love.mean(),         love.std())
print "national ice   mean=%.3f stddev=%.3f" % (national_ice.mean(), national_ice.std())
```
